Preparando o ambiente

```python
import pandas as pd
import sqlite3
```

fazendo a leitura do arquivo csv


```python
df = pd.read_csv('ca-2004-01.csv', delimiter=';', encoding='utf-8', on_bad_lines='skip')
```

```python
conn = sqlite3.connect('dados_fuel.db')
df.to_sql('dados_fuel', conn, if_exists='replace', index=False)

query = "SELECT name FROM sqlite_master WHERE type='table';"
tables = pd.read_sql(query, conn)
tables
```

imprimi a região,Estado e Municipio

```python
query = "SELECT [Regiao - Sigla], [Estado - Sigla], Municipio FROM dados_fuel;"
info_geral = pd.read_sql(query, conn)
info_geral
```

Imprimi informações sobre revenda do posto

```python
# Consultar informações sobre revenda
query = "SELECT Revenda, [CNPJ da Revenda], [Nome da Rua], [Numero Rua] FROM dados_fuel;"
info_revenda = pd.read_sql(query, conn)
info_revenda
```

Mostra algumas informações adicionais sobre o posto.

```python
query = "SELECT Complemento, Bairro, Cep, Produto, [Data da Coleta] FROM dados_fuel;"
info_adicionais = pd.read_sql(query, conn)
info_adicionais
```

Consulta valores e unidades de medidas

```python
query = "SELECT [Valor de Venda], [Valor de Compra], [Unidade de Medida] FROM dados_fuel;"
info_valores = pd.read_sql(query, conn)
info_valores
```

Consulta a informação sobre a bandeira

```python
# Consultar informações sobre bandeira
query = "SELECT Bandeira FROM dados_fuel;"
info_bandeira = pd.read_sql(query, conn)
info_bandeira
```
